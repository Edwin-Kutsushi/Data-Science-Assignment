```
import pandas as pd
```

```
# Loading our table from the respective CSV files 
with open('/content/student-mat.csv','r') as f:
    student_mat = pd.read_csv(f, index_col=0, encoding='utf-8')
# saving the urban center CSV file into a database
%sql DROP TABLE if EXISTS student_mat;
%sql PERSIST student_mat;

# Displaying the first  5 entries 
%sql SELECT * FROM student_mat LIMIT 5;
```

## Perfoming Arithmetic Operation

```
# finding the sum of the rows with numeric values
%%sql

SELECT SUM(age) FROM student_mat;
```

```
# finding the maximum age of the students
%%sql
SELECT MAX(age) FROM student_mat;
```

```
# finding the minimum age of the students
%%sql

SELECT MIN(age) FROM student_mat
```

```
# finding the total sum of the grades
%%sql

SELECT *FROM student_mat WHERE(G1+G2+G3) LIMIT 1;
```

```
# selecting certain columns
%%sql
SELECT school, G1, G2, G3
FROM student_mat DESC LIMIT 1;
```

```
# ordering in desc or asc order
%%sql

SELECT school, G1, G2, G3, age
FROM student_mat
ORDER BY age DESC LIMIT 5;
```

```
# renaming of columns
%%sql
SELECT Medu AS 'Mothers_Education'
FROM student_mat LIMIT 5;
```

```
# counting values
%%sql
SELECT COUNT(age)
FROM student_mat;
```

```
# grouping functions 
%%sql

SELECT school, COUNT(*)
FROM student_mat
GROUP BY school;
```

```
%%sql

SELECT sex, COUNT(*)
FROM student_mat
GROUP BY sex;
```

```
%%sql

SELECT MJob, COUNT(*)
FROM student_mat
GROUP BY MJob;
```

```
# checking for empty values
%%sql
SELECT age
FROM student_mat
WHERE age IS NULL;
```

```
# checking for a pattern in data
%%sql
SELECT age
FROM student_mat
WHERE age LIKE pattern;
```

```
# including rows meeting certain condition
%%sql
SELECT age
FROM student_mat
WHERE age = 20
   OR age = 22;
```

```
# ordaring our data
%%sql
SELECT age
FROM student_mat
ORDER BY age DESC LIMIT 5;
```

```
# selecting a column

%%sql
SELECT FJob 
FROM student_mat LIMIT 5;
```

```
# selecting specific columns
%%sql
SELECT * FROM student_mat
WHERE famsize='LE3' OR G1 =20;
```

```
# using the condition between
%%sql
SELECT * FROM student_mat
WHERE G3 BETWEEN 19 AND 20;
```
